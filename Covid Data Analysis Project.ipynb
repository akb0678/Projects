7. Data Aggregation:
 a. Find max value in all columns using groupby function on 
'continent'
 column
 Tip: use reset_index() after applying groupby

 (Use df_groupby dataframe for all further analysis)

a. Find max value in all columns using groupby function on 
'continent'
 column
 Tip: use reset_index() after applying groupby

```python
covid_agg= covid.copy()
covid_agg = covid_agg.groupby('continent').max().reset_index()
covid_agg
```

 b. Store the result in a new dataframe named 'df_groupby'.
 (Use df_groupby dataframe for all further analysis)

```python
df_groupby = covid_agg.copy()
df_groupby.head()
```

8. Feature Engineering :
 a. Create a new feature 'total_deaths_to_total_cases' by ratio of
 'total_deaths' column to 'total_cases'

a. Perform Univariate analysis on 'gdp_per_capita' column by 
plotting

```python
plt.figure(figsize=(10,10))
sns.set()
sns.distplot(x=df_groupby['gdp_per_capita'],label="GDP_Per_Capita",color="blue", bins=25)
plt.title('Histogram of GDP_Per_Capita')
plt.legend()
```

```python
# Approach Without using df_groupby
plt.figure(figsize=(10,10))
sns.set()
sns.distplot(x=covid['gdp_per_capita'],label="GDP_Per_Capita",color="blue", bins=25)
plt.title('Histogram of GDP_Per_Capita')
plt.legend()
```

b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

```python
# Approach by using df_groupby
plt.figure(figsize=(12,9))
sns.set()
sns.scatterplot(x=df_groupby['total_cases'], y=df_groupby['gdp_per_capita'],s=200, edgecolor = 'g', linewidth = 10)
plt.xlabel('Total_Cases')
plt.ylabel('GDP_Per_Capita')
plt.title('Total Cases Upon GDP_Per_Capita')
plt.show()
```

```python
# Approach without using df_groupby
plt.figure(figsize=(12,9))
sns.set()
sns.scatterplot(x=covid['total_cases'], y=covid['gdp_per_capita'],s=200, edgecolor = 'g', linewidth = 10)
plt.xlabel('Total_Cases')
plt.ylabel('GDP_Per_Capita')
plt.title('Total Cases Upon GDP_Per_Capita')
plt.show()
```

 c. Plot Pairplot on df_groupby dataset.

 Tip : using kind='bar' in seaborn catplot

```python
# approach 1
sns.set()
plt.figure(figsize=(16,9))
sns.catplot(x='continent',y='total_cases',data=df_groupby,kind='strip')
plt.xlabel('Cotinent')
plt.ylabel('Total Cases')
plt.title('Total Cases on Different Continent')
plt.show()
```
